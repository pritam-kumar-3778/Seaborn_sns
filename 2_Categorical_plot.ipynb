```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
tips = sns.load_dataset("tips") # tips dataset type is - pandas dataframe
print(tips.head())
```

```python
# Bar plot
sns.barplot(
    data = tips,
    x = "day",
    y = "tip",
    hue = "sex"
)
```

```python
# Box plot
sns.boxplot(
    data = tips,
    x = "day",
    y = "tip",
    hue = "sex"
)
```
