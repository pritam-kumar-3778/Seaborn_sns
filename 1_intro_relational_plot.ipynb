```python
tips = sns.load_dataset("tips") # tips dataset type is - pandas dataframe
print(tips.head())
```

```python
# Show the relation between total)bill and tip
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip"
)
```

```python
# Highlight the smoker and non smoker
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker"
)
```

```python
# Highlight the smoker and non smoker in different style
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    style="smoker"
)
```

```python
# Show the above data in line plot
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="line",
)
```
