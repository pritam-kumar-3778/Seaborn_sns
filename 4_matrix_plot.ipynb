```python
# matrix plot
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
flights = sns.load_dataset("flights") # tips dataset type is - pandas dataframe
print(flights.head())
```

```python
flights.pivot(index="month", columns="year", values="passengers")
```

```python
# heatmap
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

sns.heatmap(
    flights_pivot,
    cmap = "coolwarm" # Optional
)

plt.title("Passenges heatmap")
plt.tight_layout()
```

```python
# Add value to each cell to know how many pasengers travel
sns.heatmap(
    flights_pivot,
    cmap = "coolwarm", # Optional
    annot= True,
    fmt = "d" # for only integer value
)

plt.title("Passenges heatmap")
plt.tight_layout()
```
